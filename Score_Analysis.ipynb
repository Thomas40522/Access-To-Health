```python
import pandas as pd
```

```python
matches = pd.read_csv("WorldCupMatches.csv")
matches.loc[:, ['Home_Team_Name', 'Home_Team_Goals', 'Away_Team_Goals', 'Away_Team_Name']]
```

```python
matches.loc[matches.Home_Team_Name == 'France', ['Home_Team_Name', 'Home_Team_Goals', 'Away_Team_Goals', 'Away_Team_Name']]
```

```python
chance = pd.DataFrame({
    'Country': [],
    'Chance': []
})
chance.loc[chance.Country == 'France', 'Chance'].sum()
```

```python
for i in matches.Home_Team_Name:
    if chance.loc[chance.Country == i, 'Chance'].sum() == 0:
        teamGoal = matches.loc[matches.Home_Team_Name == i, 'Home_Team_Goals'].sum()
        awayGoal = matches.loc[matches.Home_Team_Name == i, 'Away_Team_Goals'].sum()
        chance.loc[len(chance.index)] = [i, teamGoal/(teamGoal + awayGoal)] 
```

```python
chance
```

```python
chance.to_csv('pScore.csv', index=False)
```
